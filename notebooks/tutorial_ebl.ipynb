# Extragalactic use case: constrain an EBL scale factor
In this tutorial we will tackle an extragalactic physical use case with the help of Gammapy and the Sherpa library. 

We will use Gammapy to simulate data for two blazars at different redshift by taking into account the absorption due to the extragalactic background light (EBL, see XXX for a discussion). We will use a public (and outdated) set of instrument response functions (IRFs) describing the CTA detector. We will use Sherpa to adjust the intrinsic spectra of the blazars and an EBL scale factor, for a given EBL model, with a jointly fit procedure.

The following Gammapy classes will be used to simulate the data:
* [gammapy.scripts.CTAPerf](http://docs.gammapy.org/en/latest/api/gammapy.scripts.CTAPerf.html)
* [gammapy.scripts.CTAObservationSimulation](http://docs.gammapy.org/en/latest/api/gammapy.scripts.CTAObservationSimulation.html)
* [gammapy.spectrum.Absorption](http://docs.gammapy.org/en/latest/api/gammapy.spectrum.models.Absorption.html)
* [gammapy.spectrum.PowerLaw](http://docs.gammapy.org/en/latest/api/gammapy.spectrum.models.PowerLaw.html)

The following Sherpa classes will be used to fit the data:
* [sherpa.astro.datastack](http://cxc.harvard.edu/sherpa/ahelp/datastack.html)
* [sherpa.models](http://cxc.harvard.edu/sherpa/models/)

## Setup
As usual, we'll start with some setup...

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import astropy.units as u

from gammapy.spectrum.models import AbsorbedSpectralModel, PowerLaw, Absorption
from gammapy.scripts.cta_utils import (CTAObservationSimulation,
                                       ObservationParameters, Target)
from gammapy.scripts import CTAPerf

import sherpa.astro.datastack as sh
from sherpa.models import ArithmeticModel
from sherpa.models import Parameter as SherpaParameter
```

## Gammapy simulations

### Instrument response functions
The CTA performance for point-like sources can be load by using the [CTAPerf](http://docs.gammapy.org/en/latest/api/gammapy.scripts.CTAPerf.html) class

```python
# CTA performance (point-like responses from production 2)
irf_dir = '$GAMMAPY_EXTRA/datasets/cta/perf_prod2/point_like_non_smoothed/'
irf_file = 'South_5h.fits.gz'  # For not too faint/bright source
perf = CTAPerf.read(irf_dir + irf_file)
```

The expected numbers of events seen by CTA will be simulated with this set of IRF. The performance attributs, effective area as a function of true energy, background rate as a function of reconstructed energy and matrix of energy dispersion can be accessed via the object and can be drawn with `peek` function.

```python
print(perf.aeff.data.data[:10])
print(perf.aeff.energy.log_center()[:10])
perf.peek()
```

### EBL absorption
The absorption of very high energy gamma-rays will be take into account with the [Dominguez+2011](http://adsabs.harvard.edu/abs/2011MNRAS.410.2556D) modelisation. In Gammapy we can access the EBL models via the [Absorption](http://docs.gammapy.org/en/latest/api/gammapy.spectrum.models.Absorption.html) class

```python
absorption = Absorption.read_builtin('dominguez')

redshift = 0.5
table = absorption.table_model(parameter=redshift)
ax = plt.gca()
table.plot(ax=ax, energy_range=[0.03, 10] * u.TeV)
ax.set_yscale('log')
ax.set_ylim([1.e-4, 2.])
ax.set_ylabel(r'Absorption coefficient [$\exp{(-\tau(E))}$]')
ax.set_title('EBL model (z=' + str(redshift) + ')')
ax.grid(which='both')
```

### Spectral models
To simulate AGN data, wel will make the assumptions that the intrinsic spectrum of each blazar is modelised by a simple power law.

```python
index = 2.
amplitude = 2.e-9 * u.Unit('1 / (cm2 s TeV)')  # ~PKS 2155-304 @100 GeV in 3FHL
reference = 0.1 * u.TeV
in_pwl = PowerLaw(index=index, amplitude=amplitude, reference=reference)
```

Finally, we define for each blazar a spectral model absorbed by EBL. The two blazars are respetively set at a distance of z=0.25 and z=0.75 .

```python
z_1 = 0.25
abs_model_1 = AbsorbedSpectralModel(spectral_model=in_pwl, absorption=absorption, parameter=z_1)

z_2 = 0.75
abs_model_2 = AbsorbedSpectralModel(spectral_model=in_pwl, absorption=absorption, parameter=z_2)

ax = plt.gca()
energy_range = [0.03, 5.] * u.TeV
abs_model_1.plot(energy_range=energy_range, label='z={:.2f}'.format(z_1))
abs_model_2.plot(energy_range=energy_range, label='z={:.2f}'.format(z_2))
plt.legend(loc='best')
```

### Simulations
We have now everything to simulate observations for each blazar. Simulations corresponding to point-like observations are done with the [CTAObservationSimulation](http://docs.gammapy.org/en/latest/api/gammapy.scripts.CTAObservationSimulation.html) class. The number of expected counts in the ON and the OFF regions are simulated for a given exposure (or livetime), ON/OFF normalisation (typically, 0.2) and a given energy range.

```python
# We define the observation parameters
livetime = 50 * u.h
alpha = 0.2 * u.Unit('')
emin, emax = 0.03 * u.TeV, 5 * u.TeV
obs_param = ObservationParameters(alpha=alpha, livetime=livetime, emin=emin, emax=emax)

# Simulation of blazar 1
simu_1 = CTAObservationSimulation.simulate_obs(perf=perf,
                                               target=Target(name='blazar_1', model=abs_model_1),
                                               obs_param=obs_param,
                                               obs_id='1')

# Simulation of blazar 2
simu_2 = CTAObservationSimulation.simulate_obs(perf=perf,
                                               target=Target(name='blazar_2', model=abs_model_2),
                                               obs_param=obs_param,
                                               obs_id='2')
```

We can then have a look at the distribution for one of the osbervation with the `peek` method.

```python
simu_1.peek()
```

Finally, we store the data in OGIP format in order to be able to use the Sherpa library.

```python
simu_1.write(outdir='./data_ebl/', use_sherpa=True)
simu_2.write(outdir='./data_ebl/', use_sherpa=True)
```

## Sherpa fit
The sherpa library is really usefull when dealing with complex user model. Here, we want to adjust two intrinsic models with a commun parameter, the EBL scale factor $\alpha$, on two independant datasets, namely obs_1 and obs_2: 
$$\phi(\alpha, p_1, p_2) = \phi_{\text{int}}(E,p_1) \times \exp(-\alpha \tau(E,z)) + \phi_{\text{int}}(E,p_2) \times \exp(-\alpha \tau(E,z))$$
where p1 and p2 are the intrinsic spectra of the power law models.

### Load the simulated data
First, we load the data we simulated in the Sherpa framework

```python
sh.load_data(id='obs_1', filename='./data_ebl/pha_obs1.fits')
sh.load_data(id='obs_2', filename='./data_ebl/pha_obs2.fits')
```

### Sherpa models
Second, we define a user model using the `ArithmeticModel` class from Sherpa to deal with the absorption and the $\alpha$ parameter $\exp[-\alpha * \tau(E,Z)]$.

```python
class Ebl(ArithmeticModel):
    """
    Sherpa model definition for EBL absorption.

    Read table in gammapy-extra and create an interpolator for exp[-alpha * tau(E,Z)]
    """

    def __init__(self,
                 name='EBL',
                 filename='dominguez'):
        # Create EBL data array the same way it is done in Gammapy
        absorption = Absorption.read_builtin(filename)
        self.ebl = absorption.data

        # define redshift parameter
        par_min = absorption.data.axes[0].lo[0]
        par_max = absorption.data.axes[0].hi[-1]
        self.redshift = SherpaParameter(name, 'z', par_min.value, frozen=True, min=par_min.value, max=par_max.value)

        # define parameter alpha
        self.alpha = SherpaParameter(name, 'alpha', 1., frozen=True, min=0, max=3)

        ArithmeticModel.__init__(self, name, (self.redshift, self.alpha))

    def calc(self, p, xlo, xhi=None, *args, **kwargs):
        """
        Parameters
        ----------
        p :   list of ordered parameter values.
        x :   ndarray of domain values
        xhi : ndarray of upper bin edges.

        Returns
        -------
        ndarray of calculated function values.
        """

        z = p[0]
        alpha = p[1]
        energies = xlo * u.keV  # Cause we're using sherpa

        if xhi is not None:
            energies += xhi * u.keV  # Cause we're using sherpa
            energies *= 0.5

        return np.power(self.ebl.evaluate(parameter=z, energy=energies), alpha)
```

We then add the new model to the sherpa models library and define the models we want to adjust. 

```python
# Add sherpa model
sh.add_model(Ebl)

# define model for blazar 1
sherpa_model_1 = 'powlaw1d.pwl_1'+ '*ebl.ebl_1'
sh.set_source('obs_1', sherpa_model_1)  # dataset is linked to this model
pwl_1.ref = reference.to('keV')  # sherpa works in keV
pwl_1.ampl = amplitude.to('1 / (cm2 s keV)')  # sherpa works in keV
ebl_1.redshift = 0.25

# define model for blazar 2
sherpa_model_2 = 'powlaw1d.pwl_2'+ '*ebl.ebl_2'
sh.set_source('obs_2', sherpa_model_2)  # dataset is linked to this model
pwl_2.ref = reference.to('keV')  # sherpa works in keV
pwl_2.ampl = amplitude.to('1 / (cm2 s keV)')  # sherpa works in keV
ebl_2.redshift = 0.75
```

Now we have to link the scale parameters to get a commun EBL scale factor

```python
ebl_1.alpha.thaw()
ebl_2.alpha.link = ebl_1.alpha
print(sh.get_model('obs_1'))
print(sh.get_model('obs_2'))
```

### Fit
To fit the EBL scale we use the cash statistics, the level of background is estimated during the maximisation of the likelihood function.

```python
# Fit with Cash statistics
sh.set_stat('WStat')
sh.fit()

# # Compute confidence intervals
sh.conf()
```

We can then retrieve the fitted parameters and plot the reconstructed EBL scale factor

```python
# Access to fitted parameters
res = sh.get_conf_results()
pvals1 = zip(res.parvals, res.parmins, res.parmaxes)
pvals2 = [(v, l, h) for (v, l, h) in pvals1]
fparams = dict(zip(res.parnames, pvals2))

# Retrieve parameter and its errors
alpha = fparams['ebl_1.alpha'][0]
alpha_errmin = fparams['ebl_1.alpha'][1]
alpha_errmax = fparams['ebl_1.alpha'][2]
alpha_err = [alpha_errmin, alpha_errmax]

# Drawing
ax = plt.gca()
ax.clear()
ax.errorbar(x=[0.5], y=[alpha], xerr=[0.5], yerr=[alpha_err], c='b', fmt='o', label=r'$\alpha$ reco')
ax.plot([0.,1.],[1.,1.], c='g', label=r'$\alpha$ injected')
ax.set_xlabel('Redshift', fontdict=dict(weight='bold'))
ax.set_ylabel('EBL scale', fontdict=dict(weight='bold'))
ax.set_ylim([0.9, 1.1])
plt.legend(loc='best', numpoints=1)
```
